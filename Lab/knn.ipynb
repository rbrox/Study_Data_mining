````python
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
X = pd.DataFrame(iris.data, columns= iris.feature_names)
y = pd.DataFrame(iris.target)

data = pd.DataFrame(iris.data, columns=iris.feature_names)
data['target'] = pd.Series(iris.target)


data.head(200)
````

#### Insighting the data
The data above seems to be very organised, it would be a good idea to shuffle the data before splitting it into training and testing sets. This is to avoid any bias in the data.



````python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

X_train, X_test, y_train, y_test = train_test_split(data[iris.feature_names], data['target'],test_size=0.9, random_state=42)
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)

knn.score(X_test, y_test)
````
